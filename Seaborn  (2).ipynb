```python
import seaborn as sns
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

# Displot

```python
sns.distplot(tips['total_bill'])
```

```python
sns.displot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=15)
```

#                                            # JOINT PLOT

```python
Meaning:-Jointplot allows you to basically match up two displots for bivariate data.
```

```python
KIND OF PARAMETER:-
    "scatter"
    "reg"
    "resid"
    "kde"
    "hex"
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='resid')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

#      Pair Plot     

```python
 Meaning:- Pair plot will plot pairwise relationships across an entire dataframe(for numerical columns) and supports a color hue argument
(for categorical columns).
```

```python
sns.pairplot(tips)
```

```python

sns.pairplot(tips,hue='sex',palette='deep')
#tips.head()
```

```python
sns.pairplot(tips,hue='sex',palette='brg')
#tips.head()
```

#   Rug Plot

```python
sns.rugplot(tips['total_bill'])
```

#   #KDE PLOT(Kernel Density Estimation)

```python
sns.kdeplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['tip'])
sns.rugplot(tips['tip'])
```

#   Heat Map

```python
tips.head()
```

```python
#Matrix form for correlation data
tips.corr(numeric_only=True)
```

```python
sns.heatmap(tips.corr(numeric_only=True),cmap='coolwarm',annot=True)
```

#    Cluster Map

```python
Clustermap uses hierarchal clustering to produce a clustered version of the heatmap
```

```python
sns.clustermap(tips.corr(numeric_only=True))
```

#   Barplot & Countplot

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.boxplot(x="day",y="total_bill",data=tips,palette='rainbow')
```

```python
sns.boxplot(x="day",y="total_bill",hue="smoker",data=tips,palette="coolwarm")
```

#     # Stripplot & swarmplot

```python
sns.stripplot(x="day",y="total_bill",data=tips)
```

```python
sns.swarmplot(x="day",y="total_bill",data=tips)
```
